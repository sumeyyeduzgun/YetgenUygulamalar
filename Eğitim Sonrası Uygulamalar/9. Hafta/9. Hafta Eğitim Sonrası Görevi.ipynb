# Problem 1

Kumanda isminde bir sınıf oluşturarak aşağıdaki işlemleri yapmasını sağlayan fonksiyonları sınıf içerisinde yazalım.
1. Tv Aç
2. Tv Kapat
3. Ses Ayarları
4. Kanal Ekle
5. Kanal Sayısını Öğrenme
6. Rastgele Kanala Geçme
7. Televizyon Bilgileri


Kumandayı kapatmak için q tuşuna basılmalı ve diğer durumlarda hangi işlemi yapmak istediği sorulmalıdır. Her işlemde gerekli fonksiyonunun sınıf içerisinden çağırılması gerekmektedir.

random ve time kütüphanesini araştırarak kullanabilirsiniz.

Kendi istediğiniz özellikleri eklemekte serbetsiniz.

```python
# 1.Problemin Çözümünü Buraya Yazınız
import random
import time

class Kumanda:
    def __init__(self, acik_mi = False, ses = 0, kanal_listesi = [], suanki_kanal = "", parlaklik = 0):
        self.acik_mi = acik_mi
        self.ses = ses
        self.kanal_listesi = kanal_listesi
        self.suanki_kanal = suanki_kanal
        self.parlaklik = parlaklik

    def tv_ac(self):
        if self.acik_mi:
            print("Televizyon zaten açık.")
        else:
            self.acik_mi = True
            print("Televizyon açıldı.")
        
    def tv_kapat(self):
        if self.acik_mi:
            self.acik_mi = False
            print("Televizyon kapandı.")
        else:
            print("Televizyon zaten kapalı.")
            
    def ses_ayari(self):
        while True:
            cevap = input("Sesi azalt: '<'\nSesi artır: '>'\nÇıkış: 'q'")
            if cevap == "<":
                if self.ses != 0:
                    self.ses -= 1
                    print("Ses:", self.ses)
            elif cevap == ">":
                if self.ses_ayari != 50:
                    self.ses += 1
                    print("Ses:", self.ses)
            elif cevap == "q":
                print("Çıkış yapılıyor.")
                break

    def parlaklik_ayarla(self):
        while True:
            cevap = input("Parlaklık azalt: '+'\nParlaklık artır: '-'\nÇıkış: 'q'")
            if cevap == "-":
                if self.parlaklik != 0:
                    self.parlaklik -= 1
                    print("Parlaklık:", self.parlaklik)
            elif cevap == "+":
                if self.parlaklik != 50:
                    self.parlaklik += 1
                    print("Parlaklık:", self.parlaklik)
            elif cevap == "q":
                print("Çıkış yapılıyor.")
                break
        
    def kanal_ekle(self, kanal_ismi):
        print("Kanal ekleniyor.")
        time.sleep(1)
        self.kanal_listesi.append(kanal_ismi)
        print("Kanal eklendi.")
        
    def kanal_sayisi(self):
        print("Kanal sayısı:", len(self.kanal_listesi))
        
    def rastgele_kanal(self):
        rastgele = random.randint(0, len(self.kanal_listesi)-1)
        self.suanki_kanal = self.kanal_listesi[rastgele]
        print("Şuanki kanal:", self.suanki_kanal)
        
    def televizyon_bilgileri(self):
        print("Televizyon bilgileri:")
        print("Açık mı:", self.acik_mi)
        print("Ses:", self.ses)
        print("Kanal listesi:", self.kanal_listesi)
        print("Şuanki kanal:", self.suanki_kanal)

kumanda = Kumanda()

print("""
Televizyon Kumandası

Tv Aç
Tv Kapat
Ses Ayarları
Kanal Ekle
Kanal Sayısını Öğrenme
Rastgele Kanala Geçme
Televizyon Bilgileri
Parlaklık ayarı
Çıkmak için 'q' tuşuna basın.
""")

while True:
    islem = input("İşlemi seçiniz:")
    if islem == "q":
        print("Program sonlandırılıyor...")
        break
    elif islem == "1":
        kumanda.tv_ac()
    elif islem == "2":
        kumanda.tv_kapat()
    elif islem == "3":
        kumanda.ses_ayari()
    elif islem == "4":
        kanal_isimleri = input("Kanal isimlerini virgül ile ayırarak girin:")
        kanal_listesi = kanal_isimleri.split(",")
        for kanal in kanal_listesi:
            kumanda.kanal_ekle(kanal)
    elif islem == "5":
        kumanda.kanal_sayisi()
    elif islem == "6":
        kumanda.rastgele_kanal()
    elif islem == "7":
        kumanda.televizyon_bilgileri()
    elif islem == "8":
        kumanda.parlaklik_ayarla()
    else:
        print("Geçersiz işlem.")
```
