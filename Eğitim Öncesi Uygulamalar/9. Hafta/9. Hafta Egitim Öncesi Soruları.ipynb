# Problem 1
Vehicle sınıfından miras alan bir Bus sınıfı oluşturun. Bus.seating_capacity()'nin kapasite bağımsız değişkenine varsayılan olarak 50 değerini verin.

```
Output:
> The seating capacity of a bus is 50 passengers
```


````python
# 1.Problemin Çözümünü Buraya Yazınız
class Vehicle:
    def __init__(self, model, numara, hiz, km):
        self.model = model
        self.numara = numara
        self.hiz = hiz
        self.km = km
        print("Vehicle nesnesi türetildi.")
    def intro(self):
        print(self.model, self.numara, self.hiz, self.km)
        
        
class Bus(Vehicle):
    def __init__(self, model, numara, hiz, km, seating_capacity):
        super().__init__(model, numara, hiz, km)
        self.seating_capacity = seating_capacity
        print("Bus nesnesi çalıştı")

    def intro(self):
        print(f"The seating capacity of a bus is {self.seating_capacity} passengers.")
    
    
bus = Bus("Mercedes", "Tır", 2, 100, 50)
bus.intro()

        
    
````
